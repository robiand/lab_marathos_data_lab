```sql
-- This script is used to check various Genie answers

USE CATALOG marathos;
USE SCHEMA gold;

-- check amount of athletes per country
SELECT
    athlete_country,
    count(distinct athlete_id) as athlete_count
FROM
dim_athlete
GROUP BY
athlete_country
ORDER BY
athlete_count DESC
LIMIT 10;

-- now the results can be compared to dashboard graphs

SELECT COUNT(DISTINCT event_id) AS total_events
FROM marathos.gold.fct_results;
```
