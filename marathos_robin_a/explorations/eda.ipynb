First find the correct volume directory so the csv-file can be read

```python
VOLUME_PATH = "/Volumes/marathos/default/raw"

df = (spark.read.options(header=True, inferSchema=True)
      .csv(f"{VOLUME_PATH}/data/TWO_CENTURIES_OF_UM_RACES.csv"))
```

```python
display(df.take(8))
```

Basic EDA can now be done

```python
print("Number of rows in data:", df.count())
print("Number of columns in data:", len(df.columns))

df.printSchema() # This shows the inferred schema
```

```python
df.describe().display()
```

```python
from pyspark.sql.functions import col, sum

# Count nulls
nulls = df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns])
display(nulls)
```

```python
# Count unique events
unique_events = df.select("Event name").distinct()
print("Number of unique events:", unique_events.count())

#Extra rows written by genie:
# Count unique athletes
unique_athletes = df.select("Athlete ID").distinct()
print("Number of unique athletes:", unique_athletes.count())
# Count unique countries
unique_countries = df.select("Athlete country").distinct()
print("Number of unique countries:", unique_countries.count())
```

```python
# Runner age distribution
age_distribution = df.groupBy("Athlete age category").count().orderBy("count", ascending=False)
display(age_distribution)
```

```python
# Most represented countries
country_distribution = df.groupBy("Athlete country").count().orderBy("count", ascending=False)
display(country_distribution)
```
